```python
import sqlite3 as sq
import pandas as pd
```

```python
crime = pd.read_csv("chicago_crime_2025.csv")
conn = sq.connect("crime.db")
```

```python
crime.to_sql("chicago_crime_2025",conn,if_exists="replace",index=False)
```

```python
query = """SELECT * FROM chicago_crime_2025 LIMIT 5;"""

output = pd.read_sql(query, conn) 
output
```

```python
query = """SELECT district, description, count(*) FROM chicago_crime_2025 GROUP BY district ORDER BY district;"""

output = pd.read_sql(query, conn) 
output
```

```python
query = """
SELECT
    district,
    year,
    COUNT(*) AS n_crimes,
    -- window function: percent of total crimes per district
    1.0 * COUNT(*) / SUM(COUNT(*)) OVER (PARTITION BY district) AS pct_of_district
FROM chicago_crime_2025
GROUP BY district, year
ORDER BY district, year;
"""

output = pd.read_sql(query, conn) 
output
```
